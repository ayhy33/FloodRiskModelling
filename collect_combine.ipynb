### March Street
Eddleston Water (ts_id;55790010) (station_id;36774)

```python
import requests

BASE_URL = "https://timeseries.sepa.org.uk/KiWIS/KiWIS"
TS_ID = "55790010"
START_DATE = "2023-01-01"
END_DATE = "2023-06-30"
OUTPUT_FILE = f"{TS_ID}_{START_DATE}_{END_DATE}.csv".replace("-", "")

# Configure the API request
params = {
    "service": "kisters",
    "type": "queryServices",
    "request": "getTimeseriesValues",
    "ts_id": TS_ID,
    "from": START_DATE,
    "to": END_DATE,
    "returnfields": "Timestamp,Value,Quality Code",
    "format": "csv",  # Direct CSV output
    "period": "P15M"  # Force 15-minute intervals
}

# Make the request
response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    with open(OUTPUT_FILE, 'wb') as f:
        f.write(response.content)
    print(f"Success! Data saved to {OUTPUT_FILE}")
else:
    print(f"Error {response.status_code}: {response.text}")
```

## Peebles

```python
import pandas as pd
from zipfile import ZipFile
import os
from collections import defaultdict
from io import TextIOWrapper, StringIO

def process_standard_format_file(f):
    """Process files with the standard format (3 header rows, then data)"""
    metadata = {}
    data_lines = []
    header = None
    
    for line in TextIOWrapper(f, encoding='utf-8'):
        line = line.strip()
        if line.startswith('#Timestamp;Value'):
            header = line[1:]  # Remove # but keep as header
        elif line.startswith('#') and ';' in line:
            key, val = line[1:].split(';', 1)
            metadata[key.strip()] = val.strip()
        elif line and header:  # Only collect data after header
            data_lines.append(line)
    
    if not header or not data_lines:
        return None
    
    try:
        df = pd.read_csv(StringIO(header + '\n' + '\n'.join(data_lines)), sep=';')
        # Add metadata
        for k, v in metadata.items():
            df[k] = v
        return df
    except Exception as e:
        print(f"Error parsing file: {str(e)}")
        return None
    
with ZipFile('sepa_Peebles.zip') as z:
    # ==============================================
    # 1. Process Rainfall Data (group by ts_id)
    # ==============================================
    rainfall_files = [f for f in z.namelist() 
                    if 'rainfall/' in f.lower() and f.endswith('.csv')]
    
    # Group rainfall files by ts_id
    rainfall_groups = defaultdict(list)
    for f in rainfall_files:
        filename = os.path.basename(f)
        ts_id = filename.split('_')[0]  # Extract ts_id from filename
        rainfall_groups[ts_id].append(f)
    
    # Process each rainfall group
    rainfall_data = {}
    for ts_id, files in rainfall_groups.items():
        dfs = []
        for file in files:
            with z.open(file) as f:
                df = process_standard_format_file(f)
                if df is not None:
                    dfs.append(df)
        
        if dfs:
            combined = pd.concat(dfs, ignore_index=True)
            # Cleaning steps
            combined['Timestamp'] = pd.to_datetime(combined['Timestamp'], errors='coerce')
            combined = (combined.dropna(subset=['Timestamp', 'Value'])
                        .drop_duplicates(subset=['Timestamp'])
                        .sort_values('Timestamp'))
            rainfall_data[ts_id] = combined

    # ==============================================
    # 2. Process WaterLevel Data (by ts_id)
    # ==============================================
    waterlevel_files = [f for f in z.namelist() 
                       if 'waterlevel/' in f.lower() and f.endswith('.csv')]
    
    # Group waterlevel files by ts_id
    waterlevel_groups = defaultdict(list)
    for f in waterlevel_files:
        filename = os.path.basename(f)
        ts_id = filename.split('_')[0]  # Extract ts_id from filename
        waterlevel_groups[ts_id].append(f)
    
    # Process each waterlevel group
    waterlevel_data = {}
    for ts_id, files in waterlevel_groups.items():
        dfs = []
        for file in files:
            with z.open(file) as f:
                df = process_standard_format_file(f)
                if df is not None:
                    dfs.append(df)
        
        if dfs:
            combined = pd.concat(dfs, ignore_index=True)
            # Cleaning steps
            combined['Timestamp'] = pd.to_datetime(combined['Timestamp'], errors='coerce')
            combined = (combined.dropna(subset=['Timestamp', 'Value'])
                        .drop_duplicates(subset=['Timestamp'])
                        .sort_values('Timestamp'))
            waterlevel_data[ts_id] = combined

# ==============================================
# Save Results
# ==============================================
# Create output directory
os.makedirs('clean_Peebles', exist_ok=True)

# Save rainfall data
for ts_id, df in rainfall_data.items():
    output_path = f'clean_Peebles/rainfall_{ts_id}.csv'
    df[['Timestamp', 'Value']].to_csv(output_path, index=False)
    print(f"Saved {len(df)} rainfall records for station {ts_id} to {output_path}")

# Save waterlevel data
for ts_id, df in waterlevel_data.items():
    output_path = f'clean_Peebles/waterlevel_{ts_id}.csv'
    df[['Timestamp', 'Value']].to_csv(output_path, index=False)
    print(f"Saved {len(df)} waterlevel records for station {ts_id} to {output_path}")

print("\nProcessing complete!")
print(f"→ {len(rainfall_data)} rainfall stations processed")
print(f"→ {len(waterlevel_data)} waterlevel stations processed")
```

## Edinburgh
Combine rainfall and water level data in Edinburgh

```python
import pandas as pd
from zipfile import ZipFile
import os
from collections import defaultdict
from io import TextIOWrapper, StringIO

def process_standard_format_file(f):
    """Process files with the standard format (3 header rows, then data)"""
    metadata = {}
    data_lines = []
    header = None
    
    for line in TextIOWrapper(f, encoding='utf-8'):
        line = line.strip()
        if line.startswith('#Timestamp;Value'):
            header = line[1:]  # Remove # but keep as header
        elif line.startswith('#') and ';' in line:
            key, val = line[1:].split(';', 1)
            metadata[key.strip()] = val.strip()
        elif line and header:  # Only collect data after header
            data_lines.append(line)
    
    if not header or not data_lines:
        return None
    
    try:
        df = pd.read_csv(StringIO(header + '\n' + '\n'.join(data_lines)), sep=';')
        # Add metadata
        for k, v in metadata.items():
            df[k] = v
        return df
    except Exception as e:
        print(f"Error parsing file: {str(e)}")
        return None
    
with ZipFile('sepa_Edinburgh.zip') as z:
    # ==============================================
    # 1. Process Rainfall Data (group by ts_id)
    # ==============================================
    rainfall_files = [f for f in z.namelist() 
                    if 'rainfall/' in f.lower() and f.endswith('.csv')]
    
    # Group rainfall files by ts_id
    rainfall_groups = defaultdict(list)
    for f in rainfall_files:
        filename = os.path.basename(f)
        ts_id = filename.split('_')[0]  # Extract ts_id from filename
        rainfall_groups[ts_id].append(f)
    
    # Process each rainfall group
    rainfall_data = {}
    for ts_id, files in rainfall_groups.items():
        dfs = []
        for file in files:
            with z.open(file) as f:
                df = process_standard_format_file(f)
                if df is not None:
                    dfs.append(df)
        
        if dfs:
            combined = pd.concat(dfs, ignore_index=True)
            # Cleaning steps
            combined['Timestamp'] = pd.to_datetime(combined['Timestamp'], errors='coerce')
            combined = (combined.dropna(subset=['Timestamp', 'Value'])
                        .drop_duplicates(subset=['Timestamp'])
                        .sort_values('Timestamp'))
            rainfall_data[ts_id] = combined

    # ==============================================
    # 2. Process WaterLevel Data (by ts_id)
    # ==============================================
    waterlevel_files = [f for f in z.namelist() 
                       if 'waterlevel/' in f.lower() and f.endswith('.csv')]
    
    # Group waterlevel files by ts_id
    waterlevel_groups = defaultdict(list)
    for f in waterlevel_files:
        filename = os.path.basename(f)
        ts_id = filename.split('_')[0]  # Extract ts_id from filename
        waterlevel_groups[ts_id].append(f)
    
    # Process each waterlevel group
    waterlevel_data = {}
    for ts_id, files in waterlevel_groups.items():
        dfs = []
        for file in files:
            with z.open(file) as f:
                df = process_standard_format_file(f)
                if df is not None:
                    dfs.append(df)
        
        if dfs:
            combined = pd.concat(dfs, ignore_index=True)
            # Cleaning steps
            combined['Timestamp'] = pd.to_datetime(combined['Timestamp'], errors='coerce')
            combined = (combined.dropna(subset=['Timestamp', 'Value'])
                        .drop_duplicates(subset=['Timestamp'])
                        .sort_values('Timestamp'))
            waterlevel_data[ts_id] = combined

# ==============================================
# Save Results
# ==============================================
# Create output directory
os.makedirs('clean_Edinburgh', exist_ok=True)

# Save rainfall data
for ts_id, df in rainfall_data.items():
    output_path = f'clean_Edinburgh/rainfall_{ts_id}.csv'
    df[['Timestamp', 'Value']].to_csv(output_path, index=False)
    print(f"Saved {len(df)} rainfall records for station {ts_id} to {output_path}")

# Save waterlevel data
for ts_id, df in waterlevel_data.items():
    output_path = f'clean_Edinburgh/waterlevel_{ts_id}.csv'
    df[['Timestamp', 'Value']].to_csv(output_path, index=False)
    print(f"Saved {len(df)} waterlevel records for station {ts_id} to {output_path}")

print("\nProcessing complete!")
print(f"→ {len(rainfall_data)} rainfall stations processed")
print(f"→ {len(waterlevel_data)} waterlevel stations processed")
```
